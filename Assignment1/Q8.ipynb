```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.integrate import quad


Gpctometer=3.086*10**(25)
yrtosec=3600*24*365
Gyrtosec=10**(9)*yrtosec
H=70*3.24044*10**(-20)
c=3*10**(8)
omega_m=0.3; omega_lam=0.7
```

```python
# CASE-1: Flat Lambda Cold Dark Matter Model


# Comoving Distance
def dm(zm):
    y,_=integrate.quad(lambda z:c/(H*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, zm)
   
    return y/Gpctometer 

# Luminosity Distance
def dl(zm):
    y,_=integrate.quad(lambda z:c*(1+z)/(H*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, zm)
   
    return y/Gpctometer 

# Angular Diameter Distance
def da(zm):
    y,_=integrate.quad(lambda z:c/((1+z)*H*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, zm)
   
    return y/Gpctometer 

Z=np.linspace(0,100,100)
Dm=[];Dl=[];Da=[]

for i in range(len(Z)):
    Dm.append(dm(Z[i]))

for i in range(len(Z)):
    Dl.append(dl(Z[i]))

for i in range(len(Z)):
    Da.append(da(Z[i]))


plt.plot(Z,Dm)
plt.xlabel('Redshift z')
plt.ylabel('Comoving Distance [Gpc]')
plt.title(r'Flat Lambda Cold Dark Matter Model')
plt.show()

plt.plot(Z,Dl)
plt.xlabel('Redshift z')
plt.ylabel('Luminosity Distance [Gpc]')
plt.title(r'Flat Lambda Cold Dark Matter Model')
plt.show()

plt.plot(Z,Da)
plt.xlabel('Redshift z')
plt.ylabel('Angular Diameter Distance [Gpc]')
plt.title(r'Flat Lambda Cold Dark Matter Model')
plt.show()


```
