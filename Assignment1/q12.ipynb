```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use(['science','notebook','grid'])
```

```python
h=6.62*10**(-34)
kb=1.38*10**(-23)
T=2.7255
def B_nu(nu,z):
    b=(1/(1+z)**3)*(8*np.pi*h*nu**3)/(np.exp(h*nu/kb*T)-1)
    return b
nu=np.linspace(1e6,1e11,1000)
```

```python
plt.plot(nu,B_nu(nu,0),label='z=0')
plt.plot(nu,B_nu(nu,6),label='z=6')
plt.plot(nu,B_nu(nu,20),label='z=20')
plt.plot(nu,B_nu(nu,1090),label='z=1090')
plt.plot(nu,B_nu(nu,1e4),label='z=1e4')

plt.yscale('log')
plt.ylabel('Intensity (W/m^2)')
plt.title('Logscale blackbody spectrum with for $T_{0}=T_{CMB}$')
plt.xlabel(r'$ \nu  \longrightarrow$ ')
plt.legend()
```
