```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.integrate import quad

yrtosec=3600*24*365
Gyrtosec=10**(9)*yrtosec
H=70*3.24044*10**(-20)
omega_m=0.3; omega_lam=0.7
```

```python
# CASE-1: Flat Lambda Cold Dark Matter Model

Age_Univ1,_=integrate.quad(lambda z:1/(H*(1+z)*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, np.infty)

def t(zm):
    y,_=integrate.quad(lambda z:-1/(H*(1+z)*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, zm)
   
    return y + Age_Univ1

print(t(0)/Gyrtosec,'Gyr',t(2)/Gyrtosec,'Gyr',t(6)/Gyrtosec,'Gyr',t(1100)/Gyrtosec,'Gyr')

Z=np.linspace(0,100,100)
T=[]

for i in range(len(Z)):
    T.append(t(Z[i])/Gyrtosec)

plt.plot(Z,T)
plt.xlabel('Redshift z')
plt.ylabel('Cosmic Time [Gyr]')
plt.title(r'Flat Lambda Cold Dark Matter Model')
plt.show()

```

```python
# CASE-2: Flat Dark Energy Cold Dark Matter  Model, wde=-0.7

Age_Univ2,_=integrate.quad(lambda z:1/(H*(1+z)*(omega_m*(1+z)**(3)+omega_lam*(1+z)**(0.9))**(1/2)),0, np.infty)

def t(zm):
    y1,_=integrate.quad(lambda z:-1/(H*(1+z)*(omega_m*(1+z)**(3)+omega_lam*(1+z)**(0.9))**(1/2)),0, zm)
    return y1+Age_Univ2

print(t(0)/Gyrtosec,'Gyr',t(2)/Gyrtosec,'Gyr',t(6)/Gyrtosec,'Gyr',t(1100)/Gyrtosec,'Gyr')

Z=np.linspace(0,100,100)
T=[]

for i in range(len(Z)):
    T.append(t(Z[i])/Gyrtosec)

plt.plot(Z,T)
plt.xlabel('Redshift z')
plt.ylabel('Cosmic Time [Gyr]')
plt.title(r'Flat Dark Energy Cold Dark Matter  Model, $w_{DE}$=-0.7')
plt.show()
```
