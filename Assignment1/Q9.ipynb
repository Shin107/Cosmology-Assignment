```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.integrate import quad

Gpctometer=3.086*10**(25)
yrtosec=3600*24*365
Gyrtosec=10**(9)*yrtosec
H=70*3.24044*10**(-20)
c=3*10**(8)
omega_m=0.3; omega_lam=0.7
```

```python
import numpy as np
from astropy.cosmology import WMAP9 as cosmo
cosmo.comoving_distance(np.array([0.5, 1.0, 1.5])) 
```

```python
def dm(zm):
    y,_=integrate.quad(lambda z:c/(H*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, zm)
   
    return y 

print(dm(5))
```

```python
# CASE-1: Flat Lambda Cold Dark Matter Model


# Comoving Distance
def dm(zm):
    y,_=integrate.quad(lambda z:c/(H*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, zm)
   
    return y

# Comoving Volume
def Vm(zm):
    y,_=integrate.quad(lambda z: (4*np.pi*c*dm(z)**2)/(H*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, zm)
   
    return y/(Gpctometer**(3)) 
 

# Proper Volume
def Vp(zm):
    y,_=integrate.quad(lambda z: (4*np.pi*c*dm(z)**2)/((1+z)**(3)*H*(omega_m*(1+z)**(3)+omega_lam)**(1/2)),0, zm)
   
    return y/(Gpctometer**(3)) 
 


Z=np.linspace(0,100,100)
VM=[];VP=[]

for i in range(len(Z)):
    VM.append(Vm(Z[i]))
    VP.append(Vp(Z[i]))

plt.plot(Z,VM)
plt.xlabel('Redshift z')
plt.ylabel('Comoving volume [$Gpc^{3}$]')
plt.title(r'Flat Lambda Cold Dark Matter Model')
plt.show()

plt.plot(Z,VP)
plt.xlabel('Redshift z')
plt.ylabel('Proper Volume [$Gpc^{3}$]')
plt.title(r'Flat Lambda Cold Dark Matter Model')
plt.show()
```
