```python
import numpy as np
import matplotlib.pyplot as plt

H1=70;H2=70;H3=10
sig1=300;sig2=10;sig3=300
n=200
```

```python
# H0 = 70 km/s/Mpc, Ïƒ = 300 km/s

d1=np.random.uniform(low=1, high=100, size=n)
vp1=np.random.normal(loc=0, scale=sig1,size=n)
v1=H1*d1+vp1

plt.plot(d1,v1,'ro',markersize=1)
plt.xlabel('d [Mpc]')
plt.ylabel('v [km/s]')
plt.title(r'$H_0$ = 70 km/s/Mpc,  $\sigma$ = 300 km/s')
plt.show()
```

```python
# H0 = 70 km/s/Mpc, Ïƒ = 10 km/s

d2=np.random.uniform(low=1, high=100, size=n)
vp2=np.random.normal(loc=0, scale=sig2,size=n)
v2=H2*d2+vp2

plt.plot(d2,v2,'ro',markersize=1)
plt.xlabel('d [Mpc]')
plt.ylabel('v [km/s]')
plt.title(r'$H_0$ = 70 km/s/Mpc,  $\sigma$ = 10 km/s')
plt.show()
```
